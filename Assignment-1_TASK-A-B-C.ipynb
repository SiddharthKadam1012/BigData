<h1>Simple Matrix Multiplication</h1>

```python
def sim_matrix_multiplication(matrix_1,matrix_2):
  matrix_1_row=len(matrix_1) # Total Number of rows in Matrix 1
  matrix_1_column=len(matrix_1[0])# Total Number of columns in Matrix 1
  matrix_2_row=len(matrix_2)# Total Number of rows in Matrix 2
  matrix_2_column=len(matrix_2[0])# Total Number of columns in Matrix 2
  matrix_multi=[] #Empty List to store Result of Multiplication
  if(matrix_1_column==matrix_2_row):
    for m1_r in range(len(matrix_1)): # To iterate Rows in Matrix 1
      temp = [] # to store Values on individual rows temporaraliy
      for m2_c in range(matrix_2_column): # To iterate Columns in Matrix 2
        value=0
        for m1_c in range(matrix_1_column): # To iterate columns in Matrix 1
          value+=(matrix_1[m1_r][m1_c]*matrix_2[m1_c][m2_c])
        temp.append(value)
      matrix_multi.append(temp)
  else:
    print("Multiplication Not Possible!!")
  return matrix_multi
```

```python
# Example was taken from Sem - 1 MFDS Assignment
matrix_1=[[1,2,4,1],
          [1,0,-1,-2],
          [2,-1,3,1],
          [1,2,3,4]]
matrix_2=[[-2,2,-3,3],
          [0,1,-1,2],
          [-2,-1,0,1],
          [4,0,1,4]]
np.matrix(sim_matrix_multiplication(matrix_1,matrix_2))
```

<h1> Block Matrix   Multiplication</h1>


```python
def matrix_multiplication(matrix_1,matrix_2):
  matrix_1_row=len(matrix_1) # Total Number of rows in Matrix 1
  matrix_1_column=len(matrix_1[0])# Total Number of columns in Matrix 1
  matrix_2_row=len(matrix_2)# Total Number of rows in Matrix 2
  matrix_2_column=len(matrix_2[0])# Total Number of columns in Matrix 2
  matrix_multi=[] #Empty List to store Result of Multiplication
  if(matrix_1_column==matrix_2_row):
    for m1_r in range(len(matrix_1)): # To iterate Rows in Matrix 1
      temp = [] # to store Values on individual rows temporaraliy
      for m2_c in range(matrix_2_column):# To iterate Columns in Matrix 2
        value=0
        for m1_c in range(matrix_1_column): # To iterate columns in Matrix 1
          rec_val=[]
          if type(matrix_1[m1_r][m1_c])== np.ndarray: # If Block is there , Not Value (int) we will Recursivly Call Matrix_multiplication method
            rec_val=np.array(matrix_multiplication(matrix_1[m1_r][m1_c],matrix_2[m1_c][m2_c]))
            if type(value)!= np.ndarray: # Means when value = 0
              value=np.array(rec_val)
            else:
              value+=rec_val

          else:
            rec_val=(matrix_1[m1_r][m1_c]*matrix_2[m1_c][m2_c]) # when it is values ,not block
            value+=rec_val
        temp.append(value)
      matrix_multi.append(temp)
  else:
    print("Multiplication Not Possible!!")
  return matrix_multi
```

```python
# This API only provides 20 calls and i have use it all , so it may give error to call the API

import requests
import csv


# Function to get data from the Amazon Product API and save it to a CSV file
def get_api_data(product_name,file_name,pages=1):
   # API endpoint URL
  url = "https://amazon-product-data6.p.rapidapi.com/product-by-text"


  headers = {
    "X-RapidAPI-Key": "2471033d50msh4cef20a0ee724e4p192c16jsn0e55e68aaa3e",
	  "X-RapidAPI-Host": "amazon-product-data6.p.rapidapi.com"
  }
  for i in range(1,pages+1):
    # Parameters for the API request
    querystring = {"keyword": product_name , "page": str(i), "country": "US"}
     # Sending  a GET request to the API
    response = requests.get(url, headers=headers, params=querystring)
    #print(response.json())
    # Check if the request was successful (status code 200)
    if response.status_code == 200:
           # Extracting data from the JSON response
        data = response.json().get("data", [])

        # Check if there is any data
        if data:
            # Define the CSV file path
            csv_file_path = file_name

            # Write the data to a CSV file
            with open(csv_file_path, "a", newline="", encoding="utf-8") as csv_file:
                 # Defining the CSV fieldnames (column headers)
                fieldnames = [
                    "asin",
                    "title",
                    "url",
                    "image",
                    "price",
                    "original_price",
                    "stars",
                    "rating_count",
                    "bought_in_past_month",
                    "is_prime",
                    "is_best_seller",
                    "is_climate_pledge_friendly",
                    "is_limited_time_deal",
                    "is_sponsored",
                ]

                writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
                writer.writeheader()
                # Writing each item's data as a row in the CSV file
                for item in data:
                    writer.writerow(item)

            print("Dataset created : {csv_file_path}")
        else:
            print("No data received from the API.")
    else:
        print(f"Error: API Response Error")


product_name=input("Enter Product Name:")
pages=int(input("Enter total Number of pages:"))
file_name=input("Enter the file name ie File_name.csv :")

get_api_data(product_name,file_name,pages)
```

In this part of the assignment, you will learn how to construct a dataset by scraping a
website online. Utilize BeautifulSoup or any other suitable scraping library to extract all the
details of Amazon products and compile a dataset.

```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import time
```

```python
def web_scrap_amazon(search_keyword, num_pages):
    base_url = 'https://www.amazon.in/s'
    data = []

    for page in range(1, num_pages + 1):
        params = {'k': search_keyword, 'page': page}
        webpage = requests.get(base_url, params=params)
        soup = bs(webpage.content, 'html.parser')
           # Finding  all product containers on the page
        products=soup.find_all('div',class_='puis-card-container s-card-container s-overflow-hidden aok-relative puis-include-content-margin puis puis-v1x9xtdrqwkf2b2sy3gv3bhvhpn s-latency-cf-section puis-card-border')

        # Loop through each product container
        for product in products:

            brand_name = product.find('h2').text.strip()
            rating = product.find('span', class_='a-icon-alt').text.strip() if product.find('span', class_='a-icon-alt') else 'N/A'
            reviews = product.find('span', class_='a-size-base s-underline-text').text.strip() if product.find('span', class_='a-size-base s-underline-text') else 'N/A'
            price = product.find('span', class_='a-price-whole').text.strip() if product.find('span', class_='a-price-whole') else 'N/A'

             # Append the extracted data to the 'data' list
            data.append({'brand_name': brand_name,'rating': rating, 'reviews': reviews, 'price': price})

    # Converting the list of dictionaries to a Pandas DataFrame
    return pd.DataFrame(data)

```

```python
# Amazon Web Scrapping
search_keyword = input("Enter the product Name:")
num_pages_to_scrape = int(input("Enter the Total number pages to scrap(More the 4 pages) :"))
df = web_scrap_amazon(search_keyword, num_pages_to_scrape)
df.to_csv(search_keyword+'.csv',index=True)
```
